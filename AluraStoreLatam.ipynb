```python
import pandas as pd
from IPython.display import display

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
tienda1.head()
```

#1.**Análisis de facturación**

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.


```python
# Lista con los DataFrames y sus nombres para automatizar el análisis
tiendas = {
    'Tienda 1': tienda1,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}
```

```python
# Recorremos cada tienda para la exploración de los datos
for nombre, df in tiendas.items():

    # =============================
    # TÍTULO PRINCIPAL
    # =============================
    display(f"Análisis exploratorio - {nombre}")

    # Información general del DataFrame
    display("Información general del DataFrame:")
    display(df.info())

    # Vista previa de los datos
    display("Primeras filas del DataFrame:")
    display(df.head())

    # Verificación de valores faltantes
    display("Valores faltantes por columna:")
    display(df.isnull().sum().sort_values(ascending=False))

    #Verificación de filas duplicadas
    display("Cantidad de filas duplicadas:")
    display(df.duplicated().sum())

    # Estadísticas descriptivas de columnas numéricas
    display("Estadísticas descriptivas de las columnas numéricas:")
    display(df.describe())

    # Revisión de valores únicos por columna categórica
    display("Valores únicos por columna categórica:")
    columnas_categoricas = df.select_dtypes(include="object").columns
    for col in columnas_categoricas:
        display(f"Columna '{col}' — valores únicos: {df[col].nunique()}")

    # =============================
    # SEPARADOR ENTRE TIENDAS
    # =============================
    display("=" * 60)
    display("\n")
```

```python
def calcular_ingresos_total(tiendas):
    ingresos_total = []
    for nombre, df in tiendas.items():
      try:
        ingresos = df['Precio'].sum()
        ingresos_total.append({
            'Tienda': nombre,
            'Ingreso_total': ingresos
        })
      except KeyError:
        print(f"No se encontró la columna 'Precio' en el DataFrame de {nombre}.")
      except Exception as e:
        print(f"Ocurrió un error al procesar el DataFrame de {nombre}: {e}")
    return pd.DataFrame(ingresos_total).sort_values(by='Ingreso_total', ascending=False)
```

```python
resumen_ingresos = calcular_ingresos_total(tiendas)
resumen_ingresos
```

#2. **Ventas por categorias**

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
def calcular_categorias_vendidas(tiendas):
    lista_categorias = []
    for nombre, df in tiendas.items():
        try:
            if "Categoría del Producto" not in df.columns:
                raise KeyError("Columna 'Categoría del Producto' no encontrada.")

            categorias = (
                df["Categoría del Producto"]
                .value_counts()
                .reset_index()
            )
            categorias.columns = ["Categoría", "Cantidad vendida"]
            categorias["Tienda"] = nombre

            lista_categorias.append(categorias)

        except KeyError as e:
            print(f"Error en {nombre}: {e}")
        except Exception as e:
            print(f"Error inesperado en {nombre}: {e}")


    if lista_categorias:
        df_categorias = pd.concat(lista_categorias, ignore_index=True)
        df_categorias = df_categorias.sort_values(
            by=["Tienda", "Cantidad vendida"], ascending=[True, False]
        )
        return df_categorias
    else:
        print("No se generaron datos de categorías.")
        return pd.DataFrame()
```

```python
df_categorias = calcular_categorias_vendidas(tiendas)
print("Categorías más vendidas por tienda:")
df_categorias
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
def calcular_promedio_calificaciones(tiendas):
    calificaciones = []
    for nombre, df in tiendas.items():
        try:
            if "Calificación" not in df.columns:
                raise KeyError(f"La columna 'Calificación' no se encuentra en {nombre}.")

            if df.empty:
                raise ValueError(f"El DataFrame de {nombre} está vacío.")

            calificacion_promedio = df["Calificación"].mean()

            calificaciones.append({
                "Tienda": nombre,
                "Calificaciones promedio": round(calificacion_promedio, 2)
            })

        except KeyError as e:
            print(f"Error en {nombre}: {e}")
        except ValueError as e:
            print(f"Advertencia: {e}")
        except Exception as e:
            print(f"Error inesperado en {nombre}: {e}")


    if calificaciones:
        df_calificaciones = pd.DataFrame(calificaciones)
        df_calificaciones = df_calificaciones.sort_values(by="Calificaciones promedio", ascending=False)
        return df_calificaciones
    else:
        print("No se pudo calcular el promedio de calificaciones.")
        return pd.DataFrame()
```

```python
df_calificaciones = calcular_promedio_calificaciones(tiendas)
df_calificaciones
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
def calcular_productos_vendidos(tiendas):
    resultados = []
    for nombre, df in tiendas.items():
        try:
            if 'Producto' not in df.columns:
                raise KeyError('La columna "Producto" no se encuentra en el DataFrame.')

            if df.empty:
                raise ValueError(f'El DataFrame de {nombre} está vacío')

            productos_vendidos = df['Producto'].value_counts().reset_index()
            productos_vendidos.columns = ['Producto', 'Cantidad vendida']


            productos_mas_vendidos = productos_vendidos.head(3).copy()
            productos_menos_vendidos = productos_vendidos.tail(3).copy()

            productos_mas_vendidos ['Tipo'] = 'Más vendidos'
            productos_menos_vendidos['Tipo'] = 'Menos vendidos'
            productos_mas_vendidos['Tienda'] = nombre
            productos_menos_vendidos['Tienda'] = nombre
            resultados.extend([productos_mas_vendidos, productos_menos_vendidos])

        except KeyError as e:
            print(f'Error en {nombre}: {e}')
        except ValueError as e:
            print(f'Advertencia: {e}')
        except Exception as e:
            print(f'Error inesperado en {nombre}: {e}')

    if resultados:
        df_resultados = pd.concat(resultados, ignore_index=True)
        df_resultados = df_resultados.sort_values(by=['Tienda', 'Cantidad vendida'], ascending=[True, False])
        return df_resultados
    else:
        print('No se pudo calcular los productos más vendidos.')
        return pd.DataFrame()
```

```python
df_resultados = calcular_productos_vendidos(tiendas)
display(df_resultados)
```

```python
tiendas_unicas = df_resultados["Tienda"].unique()
```

```python
for tienda in tiendas_unicas:
    print(f"{tienda}")
    print("-" * 70)

    mas_vendidos = df_resultados[
        (df_resultados["Tienda"] == tienda) &
        (df_resultados["Tipo"] == "Más vendidos")
    ].sort_values(by="Cantidad vendida", ascending=False)

    print("Productos más vendidos:")
    display(mas_vendidos)

    menos_vendidos = df_resultados[
        (df_resultados["Tienda"] == tienda) &
        (df_resultados["Tipo"] == "Menos vendidos")
    ].sort_values(by="Cantidad vendida", ascending=True)

    print("Productos menos vendidos:")
    display(menos_vendidos)

    print("\n" + "="*70 + "\n")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def graficar_productos_vertical(df, titulo="Productos más y menos vendidos por tienda"):
    """
    Genera un gráfico de barras vertical mostrando los productos más y menos vendidos
    agrupados por tienda en la misma figura.

    Parámetros:
    - df: DataFrame que debe contener las columnas 'Tienda', 'Producto', 'Cantidad vendida' y 'Tipo'
    - titulo: Título del gráfico
    """
    try:
        # Colores según tipo
        colores = {'Más vendidos': 'green', 'Menos vendidos': 'red'}

        # Crear etiquetas combinando tienda y producto
        df = df.copy()
        df['Etiqueta'] = df['Tienda'] + " - " + df['Producto']

        # Ordenar por tienda y cantidad vendida
        df = df.sort_values(by=['Tienda', 'Cantidad vendida'], ascending=[True, False])

        # Crear figura vertical
        plt.figure(figsize=(16, 8))
        plt.bar(
            x=df['Etiqueta'],
            height=df['Cantidad vendida'],
            color=df['Tipo'].map(colores),
            edgecolor='black'
        )

        # Rotar etiquetas para mejor lectura
        plt.xticks(rotation=90, ha='right')

        # Etiquetas y título
        plt.ylabel("Cantidad vendida")
        plt.xlabel("Tienda - Producto")
        plt.title(titulo, fontsize=14, fontweight='bold')
        plt.grid(axis='y', linestyle='--', alpha=0.5)

        # Leyenda manual
        verde = mpatches.Patch(color='green', label='Más vendidos')
        rojo = mpatches.Patch(color='red', label='Menos vendidos')
        plt.legend(handles=[verde, rojo], loc='upper right')

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print("Error al generar el gráfico:", e)
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
def calcular_envio_promedio(tiendas):
  envio_promedio = []
  for nombre, df in tiendas.items():
    try:
      if 'Costo de envío' not in df.columns:
        raise KeyError('La columna "Costo de envío" no se encuentra en el DataFrame.')

      if df.empty:
        raise ValueError(f'El Dataframe de {nombre} está vacío')

      promedio_envio = df['Costo de envío'].mean()
      envio_promedio.append({
      'Tienda': f'Tienda {nombre}',
      'Costo de envío promedio': round(promedio_envio, 2)
      })
    except KeyError as e:
      print(f'Error en {nombre}: {e}')
    except ValueError as e:
      print(f'Advertencia: {e}')
    except Exception as e:
      print(f'Error inesperado en {nombre}: {e}')

  if envio_promedio:
    df_envio_promedio = pd.DataFrame(envio_promedio)
    df_envio_promedio = df_envio_promedio.sort_values(by='Costo de envío promedio', ascending=False)
    return df_envio_promedio
  else:
    print('No se pudo calcular el promedio de envío.')
```

```python
df_envio_promedio = calcular_envio_promedio(tiendas)
df_envio_promedio
```

```python
def graficar_ingresos(resumen_df, titulo="Ingreso total por tienda"):
    """
    Genera un gráfico de barras vertical mostrando los ingresos totales por tienda.

    Parámetros:
    - resumen_df: DataFrame con columnas 'Tienda' y 'Ingreso_total'
    - titulo: Título del gráfico
    """
    try:
        plt.figure(figsize=(8,5))
        plt.bar(resumen_df["Tienda"], resumen_df["Ingreso_total"], color='green')

        # Títulos y etiquetas
        plt.title(titulo, fontsize=14, fontweight='bold')
        plt.xlabel("Tienda")
        plt.ylabel("Ingreso total")

        # Mostrar valores sobre las barras
        for i, valor in enumerate(resumen_df["Ingreso_total"]):
            plt.text(i, valor + 5000, f"{valor:.0f}", ha='center', fontsize=9)

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print("Error al generar el gráfico de ingresos:", e)
```

```python
def graficar_categorias_por_tienda(df_categorias, titulo="Distribución de ventas por categoría - Todas las tiendas"):
    """
    Genera gráficos de pastel para la distribución de ventas por categoría de cada tienda.

    Parámetros:
    - df_categorias: DataFrame con columnas 'Tienda', 'Categoría' y 'Cantidad vendida'
    - titulo: Título general de la figura
    """
    try:
        n_tiendas = df_categorias["Tienda"].nunique()

        # Crear figura y subplots
        fig, axes = plt.subplots(1, n_tiendas, figsize=(6*n_tiendas,6))

        # Si solo hay una tienda, axes no es lista → lo convertimos
        if n_tiendas == 1:
            axes = [axes]

        # Iterar por cada tienda y su eje
        for ax, (tienda, datos) in zip(axes, df_categorias.groupby("Tienda")):
            ax.pie(
                datos["Cantidad vendida"],
                labels=datos["Categoría"],
                autopct="%1.1f%%",
                startangle=140,
                colors=plt.cm.Paired.colors
            )
            ax.set_title(f"{tienda}", fontsize=14)

        plt.suptitle(titulo, fontsize=16, fontweight='bold')
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    except Exception as e:
        print("Error al generar gráficos de categorías por tienda:", e)


```

```python
def graficar_calificaciones_promedio(df_calificaciones, titulo="Calificaciones promedio por tienda"):
    """
    Genera un gráfico de línea con colores según el nivel de calificación.

    Parámetros:
    - df_calificaciones: DataFrame con columnas 'Tienda' y 'Calificaciones promedio'
    - titulo: Título del gráfico
    """

    try:
        # Definir colores según el valor de la calificación
        def color_calificacion(x):
            if x < 3:
                return 'red'
            elif x < 4:
                return 'orange'
            else:
                return 'green'

        colores = df_calificaciones["Calificaciones promedio"].apply(color_calificacion)

        # Crear la figura
        plt.figure(figsize=(8, 5))
        plt.plot(
            df_calificaciones["Tienda"],
            df_calificaciones["Calificaciones promedio"],
            marker='o',
            color='gray',
            linewidth=1.5,
            alpha=0.7
        )

        # Colorear los puntos individualmente
        plt.scatter(
            df_calificaciones["Tienda"],
            df_calificaciones["Calificaciones promedio"],
            color=colores,
            s=100,
            edgecolors='black'
        )

        # Etiquetas y formato
        plt.title(titulo, fontsize=14, fontweight='bold')
        plt.xlabel("Tienda")
        plt.ylabel("Calificación promedio")
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.gca().spines[['top', 'right']].set_visible(False)

        # Mostrar valores sobre los puntos
        for i, valor in enumerate(df_calificaciones["Calificaciones promedio"]):
            plt.text(i, valor + 0.02, f"{valor:.2f}", ha='center', fontsize=9)

        # Leyenda manual
        import matplotlib.patches as mpatches
        verde = mpatches.Patch(color='green', label='> 4 (Excelente)')
        naranja = mpatches.Patch(color='orange', label='3 - 4 (Buena)')
        rojo = mpatches.Patch(color='red', label='< 3 (Baja)')
        plt.legend(handles=[verde, naranja, rojo], loc='lower right')

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print("Error al generar el gráfico de calificaciones promedio:", e)
```

```python
def graficar_envio_promedio(df_envio, titulo="Costo de envío promedio por tienda"):
    """
    Genera un gráfico de barras horizontales mostrando el costo de envío promedio por tienda.

    Parámetros:
    - df_envio: DataFrame con columnas 'Tienda' y 'Costo de envío promedio'
    - titulo: Título del gráfico
    """
    try:
        plt.figure(figsize=(8,6))

        # Barras horizontales
        plt.barh(
            df_envio["Tienda"],
            df_envio["Costo de envío promedio"],
            color='skyblue',
            edgecolor='black'
        )

        # Mostrar valores al final de cada barra
        for i, val in enumerate(df_envio["Costo de envío promedio"]):
            plt.text(val + 50, i, f"{val:.2f}", va='center', fontsize=10)

        # Etiquetas y título
        plt.xlabel("Costo de envío promedio")
        plt.ylabel("Tienda")
        plt.title(titulo, fontsize=14, fontweight='bold')
        plt.grid(axis='x', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print("Error al generar gráfico de envío promedio:", e)
```

Utilice los datos de latitud (lat) y longitud (lon) para mapear las ventas de cada tienda y analizar la distribución geográfica de los productos vendidos.

```python
import plotly.express as px

def mapa_ventas_tiendas(tiendas):
    """
    Genera un mapa interactivo con la distribución geográfica
    de las ventas de todas las tiendas usando Plotly Mapbox.

    Parámetros:
    -----------
    tiendas : dict
        Diccionario donde las claves son los nombres de las tiendas
        y los valores son los DataFrames respectivos.
    """

    try:
        # Combinar todos los DataFrames en uno solo
        df_total = pd.concat(
            [df.assign(Tienda=nombre) for nombre, df in tiendas.items()],
            ignore_index=True
        )

        # Crear el gráfico interactivo con Plotly Mapbox
        fig = px.scatter_mapbox(
            df_total,
            lat="lat",                 # Columna de latitud
            lon="lon",                 # Columna de longitud
            color="Tienda",            # Diferencia las tiendas por color
            size="Precio",             # Ventas más caras = puntos más grandes
            hover_name="Producto",     # Nombre principal en el hover
            hover_data={               # Datos adicionales al pasar el mouse
                "Lugar de Compra": True,
                "Precio": ":, .0f",     # Formato de moneda
                "Tienda": True,
                "lat": False,          # Ocultar coordenadas del tooltip
                "lon": False
            },
            zoom=4.5,                  # Nivel de zoom inicial
            center={"lat": 4.60971, "lon": -74.08175},  # Centrado en Bogotá 🇨🇴
            height=600,
            title="Distribución Geográfica de Ventas por Tienda"
        )

        # Aplicar estilo de mapa base limpio
        fig.update_layout(
            mapbox_style="carto-positron",
            margin={"r": 0, "t": 50, "l": 0, "b": 10},
            title_x=0.5  # Centra el título
        )

        # Mostrar el mapa interactivo
        fig.show()

    except Exception as e:
        print(f"Error al generar el mapa: {e}")
```

```python
mapa_ventas_tiendas(tiendas)
```

Genere gráficos de dispersión o mapas de calor (Heatmaps) para visualizar datos e identificar áreas con la mayor concentración de ventas.

```python
def mapa_calor_ventas(tiendas):
    """
    Genera un mapa de calor (heatmap) para visualizar la concentración
    de ventas según las coordenadas geográficas de todas las tiendas.

    Parámetros:
    -----------
    tiendas : dict
        Diccionario donde las claves son los nombres de las tiendas
        y los valores son los DataFrames respectivos.
    """

    try:
        # Combinar todos los DataFrames en uno solo
        df_total = pd.concat(
            [df.assign(Tienda=nombre) for nombre, df in tiendas.items()],
            ignore_index=True
        )

        # Crear mapa de calor con Plotly
        fig = px.density_mapbox(
            df_total,
            lat="lat",                # Columna de latitud
            lon="lon",                # Columna de longitud
            z="Precio",               # Intensidad del color según el precio
            radius=25,                # Radio del punto de calor
            center={"lat": 4.60971, "lon": -74.08175},  # Centrado en Bogotá 🇨🇴
            zoom=4.5,                 # Nivel de zoom inicial
            mapbox_style="carto-positron",  # Estilo limpio de mapa base
            hover_name="Lugar de Compra",   # Lugar mostrado en hover
            hover_data={"Precio": ":, .0f", "Tienda": True},
            title="Mapa de Calor de Ventas por Ubicación"
        )

        # Ajustar diseño visual
        fig.update_layout(
            margin={"r": 0, "t": 50, "l": 0, "b": 10},
            title_x=0.5
        )

        # Mostrar el mapa interactivo
        fig.show()

    except Exception as e:
        print(f"Error al generar el mapa de calor: {e}")
```

```python
mapa_calor_ventas(tiendas)
```

Explore si algunas tiendas tienen un rendimiento superior o inferior al esperado en determinadas regiones e identifique si existen patrones geográficos que puedan influir en el rendimiento de las tiendas.



```python
def rendimiento_geografico_tiendas(tiendas):
    """
    Analiza el rendimiento de las tiendas por región (lat/lon)
    y muestra un mapa geográfico con colores según desempeño.

    Parámetros:
    -----------
    tiendas : dict
        Diccionario donde las claves son los nombres de las tiendas
        y los valores son los DataFrames respectivos.
    """

    try:
        # Combinar los DataFrames y agregar columna de tienda
        df_total = pd.concat(
            [df.assign(Tienda=nombre) for nombre, df in tiendas.items()],
            ignore_index=True
        )

        # Calcular el precio promedio por tienda y ubicación
        df_rendimiento = (
            df_total.groupby(["Tienda", "Lugar de Compra", "lat", "lon"], as_index=False)
            .agg({"Precio": "mean"})
            .rename(columns={"Precio": "Precio promedio"})
        )

        # Clasificar el rendimiento
        # - Alto: precio promedio > percentil 75
        # - Medio: entre 25 y 75
        # - Bajo: < percentil 25
        q25 = df_rendimiento["Precio promedio"].quantile(0.25)
        q75 = df_rendimiento["Precio promedio"].quantile(0.75)

        def clasificar_rendimiento(precio):
            if precio < q25:
                return "Bajo rendimiento"
            elif precio > q75:
                return "Alto rendimiento"
            else:
                return "Rendimiento medio"

        df_rendimiento["Rendimiento"] = df_rendimiento["Precio promedio"].apply(clasificar_rendimiento)

        # Crear mapa con Plotly
        fig = px.scatter_mapbox(
            df_rendimiento,
            lat="lat",
            lon="lon",
            color="Rendimiento",
            size="Precio promedio",
            hover_name="Lugar de Compra",
            hover_data={
                "Tienda": True,
                "Precio promedio": ":, .0f",
                "lat": False,
                "lon": False
            },
            color_discrete_map={
                "Alto rendimiento": "green",
                "Rendimiento medio": "gold",
                "Bajo rendimiento": "red"
            },
            zoom=4.5,
            center={"lat": 4.60971, "lon": -74.08175},  # Centrado en Colombia 🇨🇴
            title="Análisis de Rendimiento Geográfico de Tiendas"
        )

        # Estilo del mapa
        fig.update_layout(
            mapbox_style="carto-positron",
            margin={"r": 0, "t": 60, "l": 0, "b": 10},
            title_x=0.5
        )

        fig.show()

    except Exception as e:
        print(f"Error al generar el análisis geográfico: {e}")
```

```python
rendimiento_geografico_tiendas(tiendas)
```
